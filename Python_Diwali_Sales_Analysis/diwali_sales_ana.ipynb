```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
!pip install chardet
import chardet
```

```python
with open(r"/content/Diwali Sales Data.csv", "rb") as f:
    encoding = chardet.detect(f.read())["encoding"]
```

```python
df = pd.read_csv(r"/content/Diwali Sales Data.csv", encoding=encoding)
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(["unnamed1","Status"],axis = 1,inplace = True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df["Amount"] = df["Amount"].astype("int")
df["Amount"].dtypes
```

```python
df.columns
```

```python
df.rename(columns= {"Marital_Status" : "Slave"})
```

```python
df.describe()
```

```python
ax = sns.countplot(x="Gender",data=df)

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen = df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Gender",y = "Amount",data=sales_gen)
```

```python
ax = sns.countplot(x="Age Group",hue = "Gender",data=df)

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen = df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y = "Amount",data=sales_gen)
```

```python
sales_state = df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by ="Orders",ascending=False).head(8)

sns.set(rc={"figure.figsize":(12,4)})
sns.barplot(data=sales_state,x = "State",y = "Orders")
```

```python
sales_state = df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by ="Amount",ascending=False).head(8)

sns.set(rc={"figure.figsize":(12,4)})
sns.barplot(data=sales_state,x = "State",y = "Amount")
```

```python
ax = sns.countplot(x="Marital_Status",data=df)
sns.set(rc={"figure.figsize":(7,5)})

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state = df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by ="Amount",ascending=False).head(8)

sns.set(rc={"figure.figsize":(10,4)})
sns.barplot(data=sales_state,x = "Marital_Status",y = "Amount",hue="Gender")
```

```python
ax = sns.countplot(x="Occupation",data=df)
sns.set(rc={"figure.figsize":(17,5)})

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state = df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by ="Amount",ascending=False).head(8)

sns.set(rc={"figure.figsize":(12,4)})
sns.barplot(data=sales_state,x = "Occupation",y = "Amount")
```

```python
ax = sns.countplot(x="Product_Category",data=df)
sns.set(rc={"figure.figsize":(17,5)})

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state = df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by ="Amount",ascending=False).head(8)

sns.set(rc={"figure.figsize":(20,6)})
sns.barplot(data=sales_state,x = "Product_Category",y = "Amount")
```
